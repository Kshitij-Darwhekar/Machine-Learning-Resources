```python
football['International Reputation'].fillna(football['International Reputation'].mean(), inplace = True)
football['Skill Moves'].fillna(football['Skill Moves'].mean(), inplace = True)
football['Weak Foot'].fillna(football['Weak Foot'].mean(), inplace = True)
football['Crossing'].fillna(football['Crossing'].mean(), inplace = True)
football['Finishing'].fillna(football['Finishing'].mean(), inplace = True)
football['HeadingAccuracy'].fillna(football['HeadingAccuracy'].mean(), inplace = True)
football['ShortPassing'].fillna(football['ShortPassing'].mean(), inplace = True)
football['Volleys'].fillna(football['Volleys'].mean(), inplace = True)
football['Dribbling'].fillna(football['Dribbling'].mean(), inplace = True)
football['Curve'].fillna(football['Curve'].mean(), inplace = True)
football['FKAccuracy'].fillna(football['FKAccuracy'].mean(), inplace = True)
football['LongPassing'].fillna(football['LongPassing'].mean(), inplace = True)
football['BallControl'].fillna(football['BallControl'].mean(), inplace = True)
football['Acceleration'].fillna(football['Acceleration'].mean(), inplace = True)
football['SprintSpeed'].fillna(football['SprintSpeed'].mean(), inplace = True)
football['Agility'].fillna(football['Agility'].mean(), inplace = True)
football['Reactions'].fillna(football['Reactions'].mean(), inplace = True)
football['Balance'].fillna(football['Balance'].mean(), inplace = True)
football['ShotPower'].fillna(football['ShotPower'].mean(), inplace = True)
football['Jumping'].fillna(football['Jumping'].mean(), inplace = True)
football['Stamina'].fillna(football['Stamina'].mean(), inplace = True)
football['Strength'].fillna(football['Strength'].mean(), inplace = True)
football['LongShots'].fillna(football['LongShots'].mean(), inplace = True)
football['Aggression'].fillna(football['Aggression'].mean(), inplace = True)
football['Interceptions'].fillna(football['Interceptions'].mean(), inplace = True)
football['Positioning'].fillna(football['Positioning'].mean(), inplace = True)
football['Vision'].fillna(football['Vision'].mean(), inplace = True)
football['Penalties'].fillna(football['Penalties'].mean(), inplace = True)
football['Composure'].fillna(football['Composure'].mean(), inplace = True)
football['Marking'].fillna(football['Marking'].mean(), inplace = True)
football['StandingTackle'].fillna(football['StandingTackle'].mean(), inplace = True)
football['SlidingTackle'].fillna(football['SlidingTackle'].mean(), inplace = True)
football['GKDiving'].fillna(football['GKDiving'].mean(), inplace = True)
football['GKHandling'].fillna(football['GKHandling'].mean(), inplace = True)
football['GKKicking'].fillna(football['GKKicking'].mean(), inplace = True)
football['GKPositioning'].fillna(football['GKPositioning'].mean(), inplace = True)
football['GKReflexes'].fillna(football['GKReflexes'].mean(), inplace = True)
```

```python
#To convert String Data of 'Value' and 'Wages' columns to float value
def str2float(euros):
  #Convert M in value column to Millions
  if euros[-1] == 'M':
    return float(euros[1:-1]) * 1000000
  #Convert K in Wage column to thousands
  elif euros[-1] == 'K':
    return float(euros[1:-1]) * 1000
  else:
    return float(euros[1:])

# Use lambda function to apply UFD str2float() for each player's market value
football['Value'] = football['Value'].apply(lambda x: str2float(x))


# Use lambda function to apply UFD str2float() for each player's current wage
football['Wage'] = football['Wage'].apply(lambda x: str2float(x))
```

```python
football[['Name','Value','Wage']].head()
```

```python
#Categorical Columns
football['Club'].fillna('No Club', inplace = True)
football['Position'].fillna('Unknown', inplace = True)
```

```python
#Function to convert skill rating at each position

def skillConverter(val):
  if type(val) == str:
    s1 = val[0:2]
    s2 = val[-1]
    val = int(s1) + int(s2)
    return val
  else:
    return val


skillColumns = ['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']

for col in skillColumns:
  football[col] = football[col].apply(skillConverter)
  football[col].fillna(0.0, inplace = True)
```

```python
football.info()
```

```python
#For Height Cleaning
def clean_height(x):
  l=[]
  s=1
  if isinstance(x,str):
    l=x.split("'")
    i=int(l[0])
    j=int(l[1])
    s=((i*12)+j)/12
  return(s)

football['Height'] = football['Height'].apply(clean_height)
football['Height'].fillna(football['Height'].mean(), inplace = True)
```

```python
#For Weight Cleaning
def clean_weight(x):
  if isinstance(x,str):
    return(x.replace('lbs',''))
  return(x)


football['Weight'] = football['Weight'].apply(clean_weight).astype('float')
football['Weight'].fillna((football['Weight'].mean()), inplace = True)
```

```python
#Keeping only Numerical Value
football = football[['Name', 'Age', 'Overall', 'Potential', 'Value', 'Wage', 'Special', 'International Reputation', 'Weak Foot', 'Skill Moves',
       'Height', 'Weight', 'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']]
```

# **Hierarchical Clustering Algorithm**

## **Generating Dendrogram**

```python
import scipy.cluster.hierarchy as sch


plt.figure(figsize=(20,16))
dendrogram = sch.dendrogram(sch.linkage(X.head(40), method='single'), leaf_font_size = 13, orientation='right')

plt.title('Dendrogram')
plt.ylabel('Players')
plt.xlabel('Euclidean Distance')
```

## **Ward Linkage**

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)



plt.figure(figsize=(14,14))
ax = plt.axes()
ax.set_facecolor('grey')
plt.xlabel('Overall')
plt.ylabel('Wage')
plt.scatter(X['Overall'], X['Wage'], c=hc.labels_)

sm.silhouette_score(X, y_hc, metric = 'euclidean', sample_size = None, random_state = None)
```

## **Complete Linkage**

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'complete')
y_hc = hc.fit_predict(X)


plt.figure(figsize=(14,14))
ax = plt.axes()
ax.set_facecolor('grey')
plt.xlabel('Overall')
plt.ylabel('Wage')

plt.scatter(X['Overall'], X['Wage'], c=hc.labels_)

sm.silhouette_score(X, y_hc, metric = 'euclidean', sample_size = None, random_state = None)
```

## **Average Linkage**

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'manhattan', linkage = 'average')
y_hc = hc.fit_predict(X)


plt.figure(figsize=(14,14))
ax = plt.axes()

ax.set_facecolor('grey')
plt.xlabel('Overall')
plt.ylabel('Wage')

plt.scatter(X['Overall'], X['Wage'], c=hc.labels_)


sm.silhouette_score(X, y_hc, metric = 'manhattan', sample_size = None, random_state = None)
```

## **Single Linkage**

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'manhattan', linkage = 'single')
y_hc = hc.fit_predict(X)

plt.figure(figsize=(14,14))
ax = plt.axes()

ax.set_facecolor('grey')
plt.xlabel('Overall')
plt.ylabel('Wage')

plt.scatter(X['Overall'], X['Wage'], c=hc.labels_)


sm.silhouette_score(X, y_hc, metric = 'manhattan', sample_size = None, random_state = None)
```
